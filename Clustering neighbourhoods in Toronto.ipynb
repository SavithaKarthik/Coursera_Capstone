<h3>Scraping the Wikipedia page for the table of postal codes of Canada</h3>

BeautifulSoup Library of Python is used for web scraping of table from the Wikipedia. The title of the webpage is printed to check if the page has been scraped successfully or not. Then the table of postal codes of Canada is printed.

```python
#Webscraping
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

<h3>Webscraping Table</h3>

```python
# create a list
table_contents=[]
table=soup.find('table')
for row in table.findAll('td'):
    #create a dictionary
    cell = {}
    # Dropping the rows where Borough is 'Not assigned'
    if row.span.text=='Not assigned':
        pass
    else:
        # Extract PostalCode, Borough and Neighborhood
        cell['PostalCode'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)

#The html table is converted to Pandas DataFrame
df=pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
```

```python
# Print the dataframe
df.head()
```

<h3>Data preprocessing and cleansing</h3>

```python

# Combining the neighbourhoods with same Postalcode
df1 = df.groupby(['PostalCode','Borough'], sort=False).agg(', '.join)
df1.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df1['Neighborhood'] = np.where(df1['Neighborhood'] == 'Not assigned',df1['Borough'], df1['Neighborhood'])

df1
```

```python
# Shape of data frame
df1.shape
```

<h3>Importing the csv file containing the latitudes and longitudes for various neighborhoods in Canada</h3>

```python
#Use csv file to create dataframe with latitude and longitude data for Canada
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

<h3>Merging the two tables for getting the Latitudes and Longitudes for various neighborhoods in Canada</h3>

```python
lat_lon.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df2 = pd.merge(df1,lat_lon,on='PostalCode')
df2.head()
```

<h2>Clustering and plotting of the neighbourhoods of Canada with Toronto as Borough</h2>

<h3>Filter the rows from the data frame which contains Toronto in their Borough.</h3>

```python
df3 = df2[df2['Borough'].str.contains('Toronto',regex=False)]
df3
```

<h3>Visualizing all the Neighbourhoods of the above data frame using Folium</h3>

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighborhood in zip(df3['Latitude'],df3['Longitude'],df3['Borough'],df3['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

<h3>Using KMeans clustering for the clustering of the neighborhoods</h3>

```python
k=5
toronto_clustering = df3.drop(['PostalCode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df3.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df3['Latitude'], df3['Longitude'], df3['Neighborhood'], df3['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
